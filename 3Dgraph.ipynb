```python
import numpy as np
import matplotlib.pyplot as plt

pure = np.linspace(1, 10, 100) # 1부터 100까지 데이터 생성
noise = np.random.normal(0, 1, 100) #평균이 0, 표준편차1, 100개의 난수 생성

#넘파이 배열간 덧셈 연산, 요소별 덧셈이 수행
signal = pure + noise

# 선 그래프를 그린다
plt.plot(signal)
plt.show()
```

```python
# 학생 4명의 3과목 성적 (국어, 영어, 수학)이 넘파이 배열에 저장 되었다고 가정하자.
import numpy as np
                    # -> axis = 1 
scores = np.array([[99, 93, 60],    #
                   [98, 82, 93],    # axis =0
                   [93, 65, 81],    # ↓
                   [78, 82, 81]])
                   

print("합 : ",scores.sum())  # 합
print("최소값 : ",scores.min())  # 최소값
print("최대값 : ",scores.max())  # 최대값
print("평균 : ",scores.mean())  # 평균
print("표준편차 : ",scores.std())  # 표준편차
print("분산 : ",scores.var())  # 분산 
print("평균2 : ",scores.mean(axis=0))  #  axis = 0  행 axis = 1 열 
print("평균2 : ",scores.mean(axis=1))  #  axis = 0  행 axis = 1 열 
print("행 : ", scores.sum(axis=0))
print("행 : ", scores.sum(axis=1))

```

```python
# 히스토그램
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(1000) # seed 설정을 안하면 그래프값이 계속 바뀜 난수지만 똑같은 난수를 갖게 해줌
numbers = np.random.normal(size=1000)

plt.figure(figsize=(4,3))
plt.hist(numbers)
plt.xlabel("value")
plt.ylabel("freq")
plt.show()

```

```python
np.random.seed(10)
n, sigma = 10, 0.8 #평균, 표쥰푠처

N1 = np.random.randn(1000)
N2 = np.random.randn(1000)

N2 = n + sigma * N2

plt.figure(figsize=(6,4))
plt.hist(N1, label= "a")
plt.hist(N2, label= "b")
plt.legend(loc = "upper right")
plt.xlabel("value")
plt.ylabel("freq")
plt.show()

```

```python
# linespace()함수를 사용하여 일정간격의 데이터를 만들고 넘파이의 sin함수에 이 데이터를 전달하여 싸인값을 얻는다.


ls = np.linspace((-2*np.pi), (2*np.pi), 100)
sin = np.sin(ls)
sin2 = sin * 1/3

plt.figure(figsize=(4, 3))
plt.xlabel("value")
plt.ylabel("freq")
plt.plot(sin)
plt.plot(sin2)

plt.show()
```
