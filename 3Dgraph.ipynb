```python
#  4 3  (, , )     .
import numpy as np
                    # -> axis = 1 
scores = np.array([[99, 93, 60],    #
                   [98, 82, 93],    # axis =0
                   [93, 65, 81],    # 
                   [78, 82, 81]])
                   

print(" : ",scores.sum())  # 
print(" : ",scores.min())  # 
print(" : ",scores.max())  # 
print(" : ",scores.mean())  # 
print(" : ",scores.std())  # 
print(" : ",scores.var())  #  
print("2 : ",scores.mean(axis=0))  #  axis = 0   axis = 1  
print("2 : ",scores.mean(axis=1))  #  axis = 0   axis = 1  
print(" : ", scores.sum(axis=0))
print(" : ", scores.sum(axis=1))

```
